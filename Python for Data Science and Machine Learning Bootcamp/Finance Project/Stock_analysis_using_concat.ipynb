```python
import pandas as pd
import yfinance as yf
import numpy as np
import seaborn as sns
import datetime
%matplotlib inline
```

```python
tickers = ["BAC","C","GS","JPM","MS","WFC"] 
```

```python
ticker1 = "BAC"
BAC = yf.download(ticker1, start="2006-01-01", end="2016-01-01")

ticker2 = "C"
C = yf.download(ticker2, start="2006-01-01", end="2016-01-01")

ticker3 = "GS"
GS = yf.download(ticker3, start="2006-01-01", end="2016-01-01")

ticker4 = "JPM"
JPM = yf.download(ticker4, start="2006-01-01", end="2016-01-01")

ticker5 = "MS"
MS = yf.download(ticker5, start="2006-01-01", end="2016-01-01")

ticker6 = "WFC"
WFC = yf.download(ticker6, start="2006-01-01", end="2016-01-01")
```

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys= tickers )
```
