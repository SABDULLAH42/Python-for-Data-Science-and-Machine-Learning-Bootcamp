### Create a pairplot using seaborn of the returns dataframe. What stock stand out to you?Can you figure out why?

```python
import pandas as pd
import yfinance as yf

# Define tickers
tickers = ["BAC", "C", "GS", "JPM", "MS", "WFC"]

# Download stock data
bank_stocks = {ticker: yf.download(ticker, start="2006-01-01", end="2016-01-01") for ticker in tickers}

# Create an empty DataFrame for returns
returns1 = pd.DataFrame()

# Compute percentage change for each stock
for tick in tickers:
    returns1[tick + ' Return'] = bank_stocks[tick]['Close'].pct_change()

# Display first few rows
print(returns1.head())
```
