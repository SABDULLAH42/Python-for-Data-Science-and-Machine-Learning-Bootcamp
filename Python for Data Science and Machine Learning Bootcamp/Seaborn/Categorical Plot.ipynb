## Categorical Plot

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x ='sex', y ='total_bill', data=tips, estimator=np.std, hue='sex')
plt.show()
```

```python
sns.countplot(x='sex', data=tips, hue='sex')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,  hue='smoker')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips,  hue='day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips , hue='day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips , hue='sex', split=True)
plt.show()
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, hue='day', jitter=True)
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='day')
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips , hue='day')
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
plt.show()
```
