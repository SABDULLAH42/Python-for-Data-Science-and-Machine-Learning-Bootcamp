```python
import pandas as pd
import scipy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow
```

```python
df = pd.read_csv('C:\\Users\\USER\\Desktop\\UDEMY\\Python for Data Science and Machine Learning Bootcamp\\Machine Learning\\Deep Learning\\DATA\\cancer_classification.csv')
```

### Train Test Split

```python
X = df.drop('benign_0__mal_1', axis=1).values

y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

### Scaling Data

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#### Creating the Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Dense, Activation, Dropout
```

```python
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode = 'min', verbose=1, patience = 25 )
```

### Creating the Tensorboard Callback

```python
log_directory = "logs\\fit"
os.makedirs(log_directory, exist_ok=True)


board = TensorBoard(log_dir = log_directory, histogram_freq = 1, 
            write_graph=True, 
            write_images = True, 
            update_freq='epoch', 
            profile_batch=2, 
            embeddings_freq=1)
    




```

**Create the Model Layers:**

```python
model = Sequential()

model.add(Dense(30, activation = 'relu'))##rectified linear unit
model.add(Dropout(0.5))

model.add(Dense(15, activation = 'relu'))
model.add(Dropout(0.5))

#BINARY CLASSIFICATION
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

### Train the Model

```python
model.fit(x = X_train, 
          y = y_train, 
          epochs = 600, 
          validation_data = (X_test, y_test), verbose=1,
         callbacks = [early_stop, board]
         )

#Do not fit more than once under the same callback log_dir
```
