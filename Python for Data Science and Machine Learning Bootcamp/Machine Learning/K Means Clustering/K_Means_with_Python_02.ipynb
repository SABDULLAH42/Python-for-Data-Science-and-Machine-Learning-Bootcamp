## K Means Cluster Creation

Now its time to create the cluster labels!

```python
from sklearn.cluster import KMeans
```

#### Create an instance of a K Means model with 2 cluster


```python
kmeans = KMeans(n_clusters=2)
```

#### Fit the model to all the data except for the Private Label

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
kmeans.cluster_centers_
```

#### Create a new column for df called 'Cluster' which is 1 for Private School, and a 0 for Public School

```python
def converter (private):
    if private == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Cluster'] = df['Private'].apply(converter)
```

```python
df.head()
```

#### Create a  confusion matrix and a classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print('\n')
print(classification_report(df['Cluster'],kmeans.labels_))
```
