Its pretty difficult to visualize high dimension data in this case we have 30 dimensions

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
```

I am going to visualize the entire 3 dimensional data set just by the 2 principal components

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c = cancer['target'], cmap = 'plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
sns.set_style(style = 'darkgrid')
plt.show()
```

```python
pca.components_
```

In this Numpy matrix array each row represent a Principal Component and each column actually relates back to original features

```python
df_comp = pd.DataFrame(pca.components_, columns = cancer['feature_names'])
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp, cmap='plasma')
```
