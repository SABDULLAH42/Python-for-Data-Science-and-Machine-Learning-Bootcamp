## Now use x,y and z arrays to recreate the plot below. Notice the xlimits and ylimits on the inserted plot

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2, .5, .4, .4])


# Large Part
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

#Insert or Smaller plot
ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2, .5, .4, .4])


# Large Part
ax.plot(x,z)
ax.set_xlabel('X')
ax.set_ylabel('Z')

#Insert
ax2.plot(x,y, color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y,color='blue',lw=3,ls='--')
axes[1].plot(x,z, color='red',lw=3)
```
